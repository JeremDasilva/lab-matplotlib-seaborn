#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED

#We first look at our data to decide which plot will be the best
for i in range(1,13):
    condition = fitbit['Months_encoded'] == i
    print(f'average steps for month {i} is {fitbit[condition].Steps.mean()}')
average_per_month = [round(fitbit[fitbit['Months_encoded'] == i].Steps.mean(),2) for i in range (1,13)]
print(average_per_month)

#I estimate that bar plot is the best:
plt.bar(range(5,17), average_per_month)
plt.xticks(range(5,17), ['May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'January', 'February', 'March', 'April'], rotation = 45)
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
#We do the same as above
#1 is week, 0 is week end
for i in range(2):
    condition = fitbit['Work_or_Weekend'] == i
    print(f'average steps for {i} is {fitbit[condition].Steps.mean()}')
avg_week_or_weekend = [round(fitbit[fitbit['Work_or_Weekend'] == i].Steps.mean(),2) for i in range (2)]
plt.bar(range(2), avg_week_or_weekend)
plt.xticks(range(2), ['Week-end', 'Week'])
plt.show()
```
